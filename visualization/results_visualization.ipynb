```python
import matplotlib.pyplot as plt

# Times New Roman 폰트 설정 (선택 사항)
from matplotlib import font_manager
font_path = '/usr/share/fonts/truetype/msttcorefonts/Times_New_Roman.ttf'
font_prop = font_manager.FontProperties(fname=font_path)
plt.rcParams["font.family"] = font_prop.get_name()

# Data (*100 적용)
baseline = 0.5395 * 100
baseline_x = 0

opacity_x = [20, 40, 60, 80]
opacity_y = [0.5582, 0.5559, 0.5632, 0.5502]
opacity_y = [y * 100 for y in opacity_y]

rotation_x = [20, 40, 60, 80]
rotation_y = [0.5251, 0.5623, 0.5550, 0.5545]
rotation_y = [y * 100 for y in rotation_y]

scale_x = [20, 40, 60, 80]
scale_y = [0.5221, 0.5418, 0.5521, 0.5324]
scale_y = [y * 100 for y in scale_y]

all_x = [20, 40, 60, 80]
all_y = [0.5442, 0.5373, 0.5321, 0.5321]
all_y = [y * 100 for y in all_y]

# Plotting
plt.figure(figsize=(6, 4))

l0 = plt.scatter([baseline_x], [baseline], color='red', s=60, label='Baseline')
l1, = plt.plot(scale_x, scale_y, linewidth=2, label='Scale (S)')
l2, = plt.plot(rotation_x, rotation_y, linewidth=2, label='Rotation (R)')
l3, = plt.plot(opacity_x, opacity_y, linewidth=2, label=r'Opacity ($\alpha$)')
l4, = plt.plot(all_x, all_y, linewidth=2, label='All')

# Y-axis ticks: 10단위 + 40~60 세부 5단위 추가
yticks = [45, 50, 55, 60]

# Axes and formatting
plt.xlabel('Pruning Ratio (%)', fontsize=16)
plt.ylabel('mIoU (%)', fontsize=16)
plt.xticks([0, 20, 40, 60, 80])
plt.yticks(yticks)
plt.grid(True, linestyle='--', alpha=0.5)

# Custom legend order
plt.legend(handles=[l0, l1, l2, l3, l4])

plt.tight_layout()
plt.show()
```
